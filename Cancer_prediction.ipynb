***Splitting the data***

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2, random_state=42)
```

***Feature engineering***

```python
#this steap is known as feature engineering
```

* normalization = bringing the values to the same scale
* technies: standardscale, minmax scale

```python
#import
#from sklearn.preprocessing import MiniMaxScaler
from sklearn.preprocessing import StandardScaler
#class
```

```python
#create object to the class
sc= StandardScaler()
```

```python
#scaling the input values
x_train=sc.fit_transform(x_train)
```

```python
x_train
```

```python
x_test=sc.transform(x_test)
#fit is used by training data set
```

```python
x_test
```

***MODEL BUILDING, TRAINING, TESTING AND EVALUATION***

*Logistic regression*

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr= LogisticRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
lr_result= lr.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
lr_acc= accuracy_score(y_test, lr_result)
print("ACcuracy=",lr_acc)

con_mat=confusion_matrix(y_test, lr_result)

import seaborn as sns
plt.figure(figsize=(8,6),dpi=100)
sns.heatmap(con_mat, annot= True,fmt='d')
plt.title("Heat map for confusion matrix in LR model")
plt.xlabel("predicted value")
plt.ylabel("Actual value")


plt.show() 
```

```python
from sklearn.tree import DecisionTreeClassifier

dt=DecisionTreeClassifier()
dt.fit(x_train, y_train)
dt_result= dt.predict(x_test)

dt_acc= accuracy_score(y_test, dt_result)
print("ACcuracy=",dt_acc)

dt_cm=confusion_matrix(y_test, dt_result)


plt.figure(figsize=(8,6),dpi=100)
sns.heatmap(dt_cm, annot= True,fmt='d')
plt.title("Heat map for confusion matrix in DT model")
plt.xlabel("predicted value")
plt.ylabel("Actual value")


plt.show() 
```

```python
from sklearn.ensemble import RandomForestClassifier

rf=RandomForestClassifier()
rf.fit(x_train, y_train)
rf_result= dt.predict(x_test)

rf_acc= accuracy_score(y_test, rf_result)
print("ACcuracy=",rf_acc)

rf_cm=confusion_matrix(y_test, rf_result)


plt.figure(figsize=(8,6),dpi=100)
sns.heatmap(rf_cm, annot= True,fmt='d')
plt.title("Heat map for confusion matrix in RF model")
plt.xlabel("predicted value")
plt.ylabel("Actual value")


plt.show() 
```

```python
from sklearn.svm import SVC

svc=SVC()
svc.fit(x_train, y_train)
svc_result= dt.predict(x_test)

svc_acc= accuracy_score(y_test, svc_result)
print("ACcuracy=",svc_acc)

svc_cm=confusion_matrix(y_test, svc_result)


plt.figure(figsize=(8,6),dpi=100)
sns.heatmap(svc_cm, annot= True,fmt='d')
plt.title("Heat map for confusion matrix in SVC model")
plt.xlabel("predicted value")
plt.ylabel("Actual value")


plt.show() 
```

```python
def cancerPrediction(radius_mean=0.2, texture_mean=0.4, perimeter_mean=0.2,area_mean=0.4, smoothness_mean=0.5, compactness_mean=0.3, concavity_mean=0.72,concave_points_mean=0.11, symmetry_mean=0.43, fractal_dimension_mean=0.23,radius_se=0.123, texture_se=0.32, perimeter_se=0.98, area_se=0.99, smoothness_se=0.43,compactness_se=0.34, concavity_se=0.12, concave_points_se=0.78, symmetry_se=0.345,fractal_dimension_se=0.345, radius_worst=0.234, texture_worst=0.67,perimeter_worst=0.45, area_worst=0.23, smoothness_worst=0.34, compactness_worst=0.54, concavity_worst=0.23, concave_points_worst=0.1,symmetry_worst=0.25, fractal_dimension_worst=0.1):
    lst=[]
    lst=lst+[radius_mean]
    lst=lst+[texture_mean]
    lst=lst+[perimeter_mean]
    lst=lst+[area_mean]
    lst=lst+[smoothness_mean]
    lst=lst+[compactness_mean]
    lst=lst+[concavity_mean]
    lst=lst+[concave_points_mean]
    lst=lst+[symmetry_mean]
    lst=lst+[fractal_dimension_mean]
    lst=lst+[radius_se]
    lst=lst+[texture_se]
    lst=lst+[perimeter_se]
    lst=lst+[area_se]
    lst=lst+[smoothness_se]
    lst=lst+[compactness_se]
    lst=lst+[concavity_se]
    lst=lst+[concave_points_se]
    lst=lst+[symmetry_se]
    lst=lst+[fractal_dimension_se]
    lst=lst+[radius_worst]
    lst=lst+[texture_worst]
    lst=lst+[perimeter_worst]
    lst=lst+[area_worst]
    lst=lst+[smoothness_worst]
    lst=lst+[compactness_worst]
    lst=lst+[concavity_worst]
    lst=lst+[concave_points_worst]
    lst=lst+[symmetry_worst]
    lst=lst+[fractal_dimension_worst]
    # lst=lst+[]
    #lst=sc.transfrom([[lst]])
    
    result=lr_model.predict([lst])
    
    
    if result==0:
        return "B"
    elif result==1:
        return "M"
    
```
